**Script for scraping match data from FB Ref with xG from top leagues**

_Does not include current season (2022-23)._

```python
## Import

import requests
from bs4 import BeautifulSoup
import pandas as pd 
import numpy as np
from datetime import date
import time

```

```python
## Loop for URL List

years = ["2017-2018","2018-2019","2019-2020","2020-2021","2021-2022"]
leagues = ["Premier-League","La-Liga","Serie-A","Bundesliga","Ligue-1"]

URL_list = []

for l in leagues:
  for y in years:

    # Conditional number per League:
    if l=="Premier-League":
        x = "9"
    elif l=="La-Liga":
        x = "12"
    elif l=="Serie-A":
        x = "11"
    elif l=="Bundesliga":
        x = "20"
    elif l=="Ligue-1":
        x = "13"
    else: 
        x="0"


    tmp_URL= "https://fbref.com/en/comps/" + x + "/" + y + "/schedule/" + y + "-" + l + "-Scores-and-Fixtures"
    URL_list.append(tmp_URL)  
```

```python
## Loop for URL List

years = ["2017-2018","2018-2019","2019-2020","2020-2021","2021-2022"]
leagues = ["Premier-League","La-Liga","Serie-A","Bundesliga","Ligue-1"]

URL_list = []

for l in leagues:
  for y in years:

    # Conditional number per League:
    if l=="Premier-League":
        x = "9"
    elif l=="La-Liga":
        x = "12"
    elif l=="Serie-A":
        x = "11"
    elif l=="Bundesliga":
        x = "20"
    elif l=="Ligue-1":
        x = "13"
    else: 
        x="0"


    tmp_URL= "https://fbref.com/en/comps/" + x + "/" + y + "/schedule/" + y + "-" + l + "-Scores-and-Fixtures"
    URL_list.append(tmp_URL)  



## Scraping Results over URL List

appended_data = [] ## Empty data frame

for i in URL_list:

  print(i)
  
  time.sleep(6) # Sleep for 6 seconds
  
  # Reading in data frame
  table_data = pd.read_html(i)
  df = table_data[0]

  print(df['Home'].unique())

  # Adding league name
  if "Premier-League" in i:
    df['League'] = "EPL"
  elif "La-Liga" in i:
    df['League'] = "LaLiga"
  elif "Serie-A" in i:
    df['League'] = "Serie A"
  elif "Bundesliga" in i:
    df['League'] = "Bundesliga"
  elif "Ligue-1" in i:
    df['League'] = "Ligue 1"
  else: 
    df['League'] = "Other"

  # Subsetting columns
 # df2 = df.iloc[:,0:15]

  df2 = df.loc[:,'Wk':'League']


  # Splitting Score colum into two columns
  df2[['Home_Score', 'Away_Score']] = df2['Score'].str.split('–', expand=True)
  # Renaming xG columns
  df2.rename(columns={'xG': 'xG_Home', 'xG.1': 'xG_Away'}, inplace=True)

  # store DataFrame in list
  appended_data.append(df2)  


all_data = pd.concat(appended_data)
all_data


```

```python
## Cleaning up data 

## Removing extra columns 
all_data1 = all_data.drop(['Score', 'Notes','Match Report'], axis=1)

## removing blank rows 
all_data1 = all_data1.replace(r'^s*$', float('NaN'), regex = True)  # Replace blanks by NaN
all_data1.head(20)                                      # Print updated data
all_data1.dropna(inplace = True)                      # Remove rows with NaN

all_data1.tail(20)
```
